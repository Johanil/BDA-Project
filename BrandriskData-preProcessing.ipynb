```python
path='./data/Brandriskdata 2000-2020.csv'
#'PunktID': str, 'E': str, 'N': str,
types={ 'PunktID': str}
```

```python
df = pd.read_csv(path, sep=';', dtype=types)
```

```python
tm = projections.make_transverse_mercator("SWEREF_99_TM")
```

```python
def swerefToLatLon(dataframe):
    for row in dataframe:
        lat, lon = tm.grid_to_geodetic(row["N"], row["E"])
        row['N'] = lat
        row['E']= lon
    return dataframe
```

```python
data = swerefToLatLon(df)
```
